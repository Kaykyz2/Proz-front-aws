```python
def calculadora(num1, num2, operacao):
    if operacao == 1:
        return num1 + num2
    elif operacao == 2:
        return num1 - num2
    elif operacao == 3:
        return num1 * num2
    elif operacao == 4:
        if num2 != 0:
            return num1 / num2
        else:
            return "Erro: Divisão por zero."
    else:
        return 0
num1 = float(input("Digite e o primeiro número: "))
num2 = float(input("Digite e o segundo número: "))

print("Escolha '1' para soma, '2' para subtração, '3' para multiplicação e '4' para divisão")

operacao = int(input("Qual a operação desejada?"))

resultado = calculadora(num1, num2, operacao)

print(f"o resultado é: {resultado}")

```
