```python
rodas = int(input("Informe a quantidade de rodas do veículo: "))
peso = float(input("Informe o peso bruto do veículo (em kg): "))
pessoas = int(input("Informe a quantidade de pessoas no veículo: "))

if rodas > 1 and rodas <= 3:
  print("O veiculo se encaixa na categoria A")
elif rodas == 4 and pessoas <= 8 and peso <= 3500:
  print("O veiculo se encaixa na categoria B")
elif rodas >= 4 and peso >= 3501 and peso <= 6000:
  print("O veiculo se encaixa na categoria C")
elif rodas >= 4 and pessoas > 8 and peso < 6000:
  print("O veiculo se encaixa na categoria D")
else:
  print("O veiculo se encaixa na cateogira E")
```
