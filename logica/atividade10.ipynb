```python
def obterDadosUsuario():
    while True:

        nome = input("Digite o seu nome completo: ")


        try:
            anoNascimento = int(input("Digite o seu ano de nascimento (entre 1922 e 2024): "))
            if anoNascimento < 1922 or anoNascimento > 2024:
                print("Erro: O ano deve estar entre 1922 e 2024.")
                continue
        except ValueError:
            print("Erro: Por favor, digite um número válido.")
            continue

        anoAtual = 2022
        idade = anoAtual - anoNascimento


        print(f"{nome}, você completou ou completará {idade} anos em 2022.")
        break

# Chama a função
obterDadosUsuario()
```
